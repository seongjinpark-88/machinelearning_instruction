# Discrimination between lateral and retroflex

**Note:** 이 `notebook`을 실행하기 이전에 다음의 개념을 설명하고 진행하여야 합니다. 
 1. *Deep* model
 2. *Wide* model
 3. Dropout
 4. One-hot encoding

`1-2`를 진행할 때, 간단한 `로지스틱 회귀 (logistic regression)`을 이용하여서 `lateral/retroflex` 분류를 실행하였습니다. 이번에는 **deep and wide** 모델을 구축하여서 분류의 성능이 어떻게 변화하는지 살펴보도록 하겠습니다. 

## Check the data

이전에 사용하였던 데이터의 구조를 다시 한 번 확인하도록 하겠습니다. 

첫 줄은 각각의 열이 어떤 값을 나타내는지를 보여주는 **header**입니다. 그 이후 내용들은 각각의 화자가 `lateral/retroflex`를 발화하였을 때 어떠한 값들이 측정되었는지를 나타냅니다.

## Define Min-Max normalization

`min-max normalization` 기능을 `python`에서 정의하도록 하겠습니다. 

```python
def minmax_scale(original_array, target_array):
    (min_f2, min_f3) = np.min(original_array, axis = 0)
    (max_f2, max_f3) = np.max(original_array, axis = 0)
    for i in range(0, len(original_array)):
        for j in range(0, len(original_array[i])):
            if (j == 0):
                target_array[i][j] = (original_array[i][j] - min_f2) / (max_f2 - min_f2)
            else:
                target_array[i][j] = (original_array[i][j] - min_f3) / (max_f3 - min_f3)
    return target_array
            
```

## Import modules

모델 구축에 필요한 모듈을 불러오도록 하겠습니다. 

```python
import numpy as np
from keras import optimizers
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.callbacks import TensorBoard
from keras.utils import np_utils
```

## Load the data and create a dataset

데이터를 불러온 이후, 입력값과 출력값을 저장하는 `<list>`를 생성하여 각각 저장하도록 하겠습니다. 이전 `notebook`에서 했던 것 처럼 데이터는 무작위로 섞도록 하겠습니다. 생성 과정은 이전과 동일합니다. 

데이터를 할당한 이후, 입력값과 출력값 첫 20개를 확인해보도록 하겠습니다. 

```python
with open("data/liquid_regression.txt", "r") as f:
    data = f.readlines()
    f.close()
    
import random

X_data = []
y_data = []

ndata = data[1:]
random.shuffle(ndata)

for i in ndata:

    line = i.strip()
    (spk, con, f2, f3) = line.split("\t")
    f2 = float(f2)
    f3 = float(f3)
    X = [f2, f3]

    if (con == "retroflex"):
        y = [0.]
    elif (con == "lateral"):
        y = [1.]
    X_data.append(X)
    y_data.append(y)

print(len(X_data))
print(X_data[0:20])
print(y_data[0:20])
```

## Do min-max normalization, and split the data into training, validation, and test dataset

전체 데이터에 `min-max normalization`을 적용한 이후, 전체 데이터를 세 개의 데이터셋(`training`, `validation`, 그리고 `test`)으로 분류하도록 하겠습니다. 

```python
X_zeros = np.zeros_like(X_data)
X_data = minmax_scale(X_data, X_zeros)

X_train = np.array(X_data[0:250])
y_train = np.array(y_data[0:250])

X_val = np.array(X_data[250:300])
y_val = np.array(y_data[250:300])

X_test = np.array(X_data[300:])
y_test = np.array(y_data[300:])
```

# Model with one hidden layer

맨 먼저 하나의 `은닉 계층 (Hidden layer)`을 지닌 모델을 만들어 보겠습니다. 이전 `notebook`과 동일하게 하나의 `은닉 계층`에 두 개의 노드를 가지고 있는 `layer`입니다. 


## Create a model with one hidden layer

각각의 계층의 파라미터 수는 **summary()** 명령어를 이용하여 확인할 수 있습니다. 모델의 구조를 설정하도록 하겠습니다. 

```python
model = Sequential()

model.add(Dense(2, input_dim = 2, activation = 'sigmoid'))
model.add(Dense(1, activation = 'sigmoid'))

adam = optimizers.Adam(lr=0.001)

model.compile(loss="binary_crossentropy", optimizer=adam, metrics = ['accuracy'])

model.summary()
```

## Training the model

모델을 학습시키고 과정을 확인하도록 하겠습니다. 

```python
tensorboard = TensorBoard(log_dir='./logs', histogram_freq = 0, 
                         write_graph=True, write_images=False)

model.fit(X_train, y_train, batch_size = 10, epochs = 100, verbose = 1, validation_data=(X_val, y_val), 
          shuffle = True, 
          callbacks=[tensorboard])
```

## Evaluation

실제 결과와 예측값을 확인해보겠습니다. 

```python
predictions = model.predict(X_test)
for j in range(0, 20):
    print(y_test[j], "\t", predictions[j][0], "\t", round(predictions[j][0]))
```

실제 `loss`와 `accuracy`가 어떻게 되는지 확인해보겠습니다. 

```python
score = model.evaluate(X_test, y_test, verbose = 0)

print("Test score:\t", score[0])
print("Test accuracy:\t", score[1])
```

모델에서는 `오버피팅 (over-fitting)`이 일어났습니다. 즉, `training` 데이터에서는 높은 accuracy를 보였지만, `validation` 및 `test` 데이터에서는 상대적으로 낮은 accuracy를 보였습니다. 

모델의 구조를 변경한 이후, 결과가 어떻게 나타나는지 살펴보겠습니다. 

# One hidden layer + Dropout layer

## Add Dropout

동일한 구조의 모델에 `Dropout` 계층을 하나 추가하겠습니다. `Dropout` 기능은 쉽게 말해 실제 활성화되는 값 중 일정 비율을 `비활성화`시켜서 `오버피팅(over-fitting)`을 방지하는 방법입니다. 대부분의 경우 활성화되는 값이 많을 경우 더 큰 성능을 발휘하지만, 이런 작은 모델에서도 성능 향상에 도움이 되는지 확인해보겠습니다. 

```python
model_drop = Sequential()

model_drop.add(Dense(2, input_dim = 2, activation = 'sigmoid'))
model_drop.add(Dropout(0.25))
model_drop.add(Dense(1, activation = 'sigmoid'))

adam = optimizers.Adam(lr=0.001)

model_drop.compile(loss="binary_crossentropy", optimizer=adam, metrics = ['accuracy'])

model_drop.summary()
```

## Train the model

구축된 모델에 같은 데이터로 학습을 시켜보겠습니다. 

```python
model_drop.fit(X_train, y_train, batch_size = 10, epochs = 100, verbose = 1, validation_data=(X_val, y_val), 
          shuffle = True, 
          callbacks=[tensorboard])
```

## Check the result



```python
score = model_drop.evaluate(X_test, y_test, verbose = 0)

print("Test score:\t", score[0])
print("Test accuracy:\t", score[1])
```

`Dropout` 기능을 추가한 경우, 모델의 정확도가 향상되었으며 `오버피팅 (over-fitting)`도 일어나지 않았음을 볼 수 있습니다. 

이제는 `Dropout` 기능 대신에 더 **깊은** 구조의 모델과, 더 **넓은** 구조의 모델을 만들어 보도록 하겠습니다. 

# Model with two hidden layers

이번에 만들어볼 모델은 더 **깊은 구조**의 모델입니다. 즉, 더 많은 `은닉 계층 (Hidden layer)`를 가지고 있는 모델입니다. 각 은닉 계층의 `node` 수는 변화시키지 않겠습니다. 우선 모델을 생성하겠습니다. 


## Create model with two hidden layers (deeper)

```python
model2 = Sequential()

model2.add(Dense(2, input_dim = 2, activation = 'sigmoid'))
model2.add(Dense(2, activation = 'sigmoid'))
model2.add(Dense(1, activation = 'sigmoid'))

adam = optimizers.Adam(lr=0.001)

model2.compile(loss="binary_crossentropy", optimizer=adam, metrics = ['accuracy'])

model2.summary()
```

모델의 구조는 앞과 거의 유사합니다. 차이가 있다면 하나였던 `은닉 계층`이 두 개로 늘었다는 것입니다. 그럼 이 모델을 학습시켜보겠습니다. 

## Train the model

```python
tensorboard = TensorBoard(log_dir='./logs', histogram_freq = 0, 
                         write_graph=True, write_images=False)

model2.fit(X_train, y_train, batch_size = 10, epochs = 100, verbose = 1, validation_data=(X_val, y_val), 
          shuffle = True, 
          callbacks=[tensorboard])
```

이전과 같은 횟수의 `epoch` 값을 설정한 경우, 결과값이 매우 낮게 나오는 것을 확인할 수 있습니다. `epoch` 수를 늘려서 결과가 어떻게 변하는지 살펴보도록 하겠습니다. 

```python
model2.fit(X_train, y_train, batch_size = 10, epochs = 500, verbose = 1, validation_data=(X_val, y_val), 
          shuffle = True, 
          callbacks=[tensorboard])
```

`epoch` 값을 높인 경우 모델의 정확도도 높아졌으며, `오버피팅 (over-fitting)`도 일어나지 않았음을 알 수 있습니다. 

## Evaluation

실제 모델의 결과값과 예측값을 비교해보겠습니다. 

```python
predictions = model2.predict(X_test)
for j in range(0, 20):
    print(y_test[j], "\t", predictions[j][0], "\t", round(predictions[j][0]))
```

```python
score = model2.evaluate(X_test, y_test, verbose = 0)

print("Test score:\t", score[0])
print("Test accuracy:\t", score[1])
```

실제 정확도도 높게 나타난 것을 확인할 수 있습니다. 그러므로, 실제로 `은닉 계층`을 추가하는 것이 모델의 성능을 향상시켰다고 볼 수 있을 것입니다. 

----------

**FYI:** 실제 모델의 성능은 어떻게 데이터를 무작위로 섞었는가, 변수들의 초기값이 무엇인가에 따라서 달라질 수 있습니다.
**FYI:** 모델의 `epoch`을 높일 경우, 대부분의 경우에서 모델이 더 좋은 성능을 보일 수 있습니다. 하지만 그럴 경우 `오버피팅 (over-fitting)`이 일어나는 경우도 많으므로, 실제 `test` 데이터에 대한 결과값만이 아니라 `training` 및 `validation` 데이터에 대한 결과도 확인하여야 합니다. 

----------

이제는 더 **넓은** 구조의 모델의 성능을 확인하도록 하겠습니다. 

# Model with one hidden layer + Wider model


**더 넓은 구조**의 모델**(wide model)**은 `은닉 계층`의 수는 같지만 그 `layer`의 `node` 수가 많은 것을 의미합니다. 과연 그렇게 모델을 변화시키는 것이 성능의 향상을 가지고 오는지 확인해보겠습니다. 

## Create a model with one hidden layer, with more nodes

```python
model3 = Sequential()

model3.add(Dense(10, input_dim = 2, activation = 'sigmoid'))
model3.add(Dense(1, activation = 'sigmoid'))

adam = optimizers.Adam(lr=0.001)

model3.compile(loss="binary_crossentropy", optimizer=adam, metrics = ['accuracy'])

model3.summary()
```

위와 같은 구조를 가진 모델을 동일한 데이터로 학습시켜보겠습니다. 

```python
model3.fit(X_train, y_train, batch_size = 10, epochs = 100, verbose = 1, validation_data=(X_val, y_val), 
          shuffle = True, 
          callbacks=[tensorboard])
```

## Evaluation

실제 결과값과 예측한 값을 비교해보도록 하겠습니다. 


```python
predictions = model3.predict(X_test)
for j in range(0, 20):
    print(y_test[j], "\t", predictions[j][0], "\t", round(predictions[j][0]))
```

```python
score = model3.evaluate(X_test, y_test, verbose = 0)

print("Test score:\t", score[0])
print("Test accuracy:\t", score[1])
```

결과를 보면, `test` 데이터에 대한 정확도는 크게 다르지 않다는 것을 확인할 수 있습니다. 하지만 같은 `epoch` 값을 주었을 때, 세 번째 모델(**wide model**)에서 가장 높은 정확도를 보인다는 것을 확인할 수 있습니다. 

## One-hot encoding

**`one-hot encoding`** 은 분류 문제(*classification*)를 해결할 때 사용하는 방법 중 하나입니다. 이전처럼 하나의 값을 산출해낸 이후, 그 값에 따라 두 개로 분류할 수도 있지만, 분류할 범주만큼의 **노드**를 생성한 이후, 각각의 노드 중, 더 큰 확률값을 가지는 노드를 선택하여 분류하는 방법도 사용할 수 있습니다. 그럴 때 사용하는 방식이 **one-hot encoding**입니다. 

이전에 /l/과 /r/을 구분할 때 /r/을 0., /l/을 1. 의 값을 할당하여 분류를 진행하였다면, 이제는 `retroflex`의 경우 `[1., 0.]`이라는 `<list>`로 값을 할당하고, `lateral`의 경우 `[0., 1.]`이라는 `<list>`로 값을 할당하도록 하겠습니다. 즉 **one-hot encoding**은 실제 분류에 해당하는 값에 **1** 이라는 값을 할당하고 나머지에는 **0**을 할당함으로써 범주를 구분하는 방법이라고 볼 수 있습니다. 

가장 높은 정확도를 보였던 세 번째 모델(wide model)에 **one-hot encoding**을 적용한 이후, 그 결과를 살펴보도록 하겠습니다. 



```python
model4 = Sequential()

model4.add(Dense(10, input_dim = 2, activation = 'sigmoid'))
model4.add(Dense(2, activation = 'softmax'))

adam = optimizers.Adam(lr=0.001)

model4.compile(loss="categorical_crossentropy", optimizer=adam, metrics = ['accuracy'])

model4.summary()
```

```python
onehot_train = np_utils.to_categorical(y_train, 2)
onehot_val = np_utils.to_categorical(y_val, 2)
onehot_test = np_utils.to_categorical(y_test, 2)

tensorboard = TensorBoard(log_dir='./logs', histogram_freq = 0, 
                         write_graph=True, write_images=False)

model4.fit(X_train, onehot_train, batch_size = 10, epochs = 100, verbose = 1, validation_data=(X_val, onehot_val), 
          shuffle = True, 
          callbacks=[tensorboard])
```

```python
score = model4.evaluate(X_test, onehot_test, verbose = 0)

print("Test score:\t", score[0])
print("Test accuracy:\t", score[1])
```

눈으로 구별하기 힘든 정도의 차이지만, 실제로 아주 약간의 성능 향상은 있었습니다. 여기서 이해하여야 할 부분은 성능이 향상되었냐 아니냐에 중점을 두기 보다는, 이러한 방법으로도 **분류 문제**를 해결할 수 있음을 이해하는 것입니다.
