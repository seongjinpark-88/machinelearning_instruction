이전 노트북에서 `OR`을 **numpy**를 이용해서 구현해보았습니다. 이번에는 같은 모델을 `keras`를 사용하여 구현해보도록 하겠습니다. 

## Build the model

먼저 입력값과 출력값을 담은 **numpy array**를 생성한 이후, 실제 모델을 구축해보도록 하겠습니다. 이전에 `로지스틱 회귀 (logistic regression)`에서 했던 것 처럼, `loss`는 *binary_crossentropy*, 성능은 *accuracy*로 평가하겠습니다. 

```python
import numpy as np

from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers

X_data = [[0., 0.],
          [0., 1.],
          [1., 0.],
          [1., 1.]]

y_data = [[0.],
          [1.],
          [1.],
          [1.]]

X_data = np.array(X_data)
y_data = np.array(y_data)

model = Sequential()
model.add(Dense(1, input_dim = 2, activation = 'sigmoid'))
adam = optimizers.Adam(lr = 0.001)
model.compile(loss='binary_crossentropy', optimizer = adam, metrics=['accuracy'])
model.summary()
```

## Training the model

실제 모델을 학습시켜서 모델이 `OR` 분류를 얼마나 잘 수행하는지 살펴보겠습니다. 

```python
model.fit(X_data, y_data, epochs = 5000)
```

## Evaluation

모델의 성능을 평가해보겠습니다. 

```python
score = model.evaluate(X_data, y_data, verbose = 1)
print("Test score:\t", score[0])
print("Test Accuracy:\t", score[1])
```

```python
print(model.predict_classes(X_data))
```

## `XOR` question

출력값을 바꾸어서 모델이 `XOR`  문제를 해결할 수 있는지 확인해보겠습니다. 

```python
y_xor = [[0.],
         [1.],
         [1.],
         [0.]]

y_xor = np.array(y_xor)

model.fit(X_data, y_xor, epochs = 5000)
```

## Evaluation

같은 구조로 만든 모델은 `XOR` 분류는 제대로 수행하지 못하는 것을 확인할 수 있습니다. 실제로 어떻게 예측하였는지 한 번 살펴보겠습니다. 

```python
print(model.predict_classes(X_data))
```

```python
score2 = model.evaluate(X_data, y_xor, verbose = 1)
print("Test score:\t", score2[0])
print("Test Accuracy:\t", score2[1])
```

단순한 **Perceptron** 모델을 이용하여서는 `XOR` 분류를 제대로 수행할 수 없다는 것을 확인하였습니다. 그렇기 때문에 `다층 퍼셉트론 (Multilayer Perceptron)`의 필요성이 대두되었고, 실제로 사용되었니다. 다음 `notebook`에서 관련 내용을 확인해보겠습니다.
